Необходимые импорты

```python
import lyricsgenius
import time
```

Токен, который нужен для выкачивания.

```python
client_access_token = '***'
```

```python
api = lyricsgenius.Genius(client_access_token, timeout=10)
```

Список исполнителей открываем.

```python
artists_list = 'Список артистов.txt'
```

```python
with open(artists_list, 'r', encoding='utf-8') as f:
  artists = [line.strip() for line in f.readlines()]
```

Начинаем выкачивать информацию. Возможно несколько случае:

1 - Исполнителя нет на сайте

2 - Есть на сайте исполнитель, но нет песен

3 - Исполнитель есть на сайте, у него есть песни

Берём только песни и исключаем метки подобные "Припев", "Куплет" и т.д. Сохраняем информацию в json файл.

Исключения мы запишем в отдельные файлы.

```python
for artist in artists:
    try:
        g_artist = api.search_artist(artist)
        if g_artist.songs:
            if len(g_artist.songs) > 0:
                lyricsgenius.remove_section_headers = True # Получить только текст песни без дополнительных меток
                lyricsgenius.skip_non_songs = True # Ппропускать треки, которые не являются песнями
                lyricsgenius.excluded_terms = ["(Remix)", "(Live)", "Remix", "mix", "Edition", 'Retouch', '(Retouch)'] # Список терминов, которые будут исключены при поиске песен
                g_artist.save_lyrics(extension='json', overwrite= True)
                time.sleep(5) # Избежать блокировки
            else:
                with open ('no_data.txt', 'a+', encoding='utf-8') as f:
                    f.write(artist + ' нет песен\n')
        else:
            with open ('no_data.txt', 'a+', encoding='utf-8') as f:
                f.write(artist + ' нет исполнителя\n')
    except:
        with open ('no_data.txt', 'a+', encoding='utf-8') as f:
            f.write(artist + ' ошибка\n')
```
